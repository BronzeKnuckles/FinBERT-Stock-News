```python

from concurrent.futures import ThreadPoolExecutor
import torch
import yfinance as yf



from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

# Create a Custom Business Day object for US Federal Holidays
us_bd = CustomBusinessDay(calendar=USFederalHolidayCalendar())

# Example function to get the next business day
def get_next_business_day(date_str):
    date = pd.to_datetime(date_str)
    next_business_day = date + us_bd
    return next_business_day





def get_stock_price(ticker, date):
    stock = yf.Ticker(ticker)
    start_date = pd.to_datetime(date)
    next_business_day = get_next_business_day(start_date)
    

    # Convert back to string format for yfinance
    start_date = start_date.strftime('%Y-%m-%d')
    end_date = next_business_day.strftime('%Y-%m-%d')

    hist = stock.history(start=start_date, end=end_date)
    if not hist.empty:
        return hist['Close'][0]  
    else:
        return None
def fetch_prices_for_dataframe(df):
    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = [executor.submit(get_stock_price, row['stock'], row['date']) for index, row in df.iterrows()]
        prices = [future.result() for future in futures]
    return prices


df['stock_price'] = fetch_prices_for_dataframe(df)

```

```python
print(f'Null Values : {df["stock_price"].isnull().sum()}')
df = df.dropna()
print(f'Null Values after dropping: {df["stock_price"].isnull().sum()}')
```

```python
from scipy.special import softmax
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
def sentiment_score(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    scores = outputs.logits[0].detach().numpy()
    scores = softmax(scores)
    return scores






def fetch_sentiments_for_dataframe(df, model, tokenizer):
    with ThreadPoolExecutor(max_workers=5) as executor:  
        futures = [executor.submit(sentiment_score, row['headline']) for index, row in df.iterrows()]
        sentiments = [future.result() for future in futures]
    return sentiments


df['sentiment_scores'] = fetch_sentiments_for_dataframe(df, model, tokenizer)
```

```python
df.head(10)
```

```python
df['prev_stock_price'] = df.groupby('stock')['stock_price'].shift(1)
df['price_change_percent'] = ((df['stock_price'] - df['prev_stock_price']) / df['prev_stock_price']) * 100
def categorize_change(x):
    if x > 0.2:  # Example threshold for increase
        return 'increase'
    elif x < -0.2:  # Example threshold for decrease
        return 'decrease'
    else:
        return 'stable'

df['price_change_category'] = df['price_change_percent'].apply(categorize_change)
# Example heuristic: positive sentiment leads to price increase, etc.
def predict_price_change(sentiment_score):

    positive_score, negative_score, neutral_score = sentiment_score

    threshold = 0.75
    if positive_score > threshold:
        return 'increase'
    elif negative_score > threshold:
        return 'decrease'
    else:
        return 'stable'


df['predicted_change'] = df['sentiment_scores'].apply(predict_price_change)


df_no_nan = df.dropna(subset=['prev_stock_price'])


# Calculate accuracy and other metrics
from sklearn.metrics import accuracy_score, classification_report



def get_Metrics(df):

    accuracy = accuracy_score(df['price_change_category'], df['predicted_change'])
    report = classification_report(df['price_change_category'], df['predicted_change'])

    #print(accuracy)
    print(report)



print(f"Result for df [no null]")
get_Metrics(df_no_nan)





filtered_df = df[df['price_change_percent'].abs() >= 0.2]
print(f"Result for filtered df [over -/+.2% ]")
get_Metrics(filtered_df)









```
